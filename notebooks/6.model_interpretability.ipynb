```python
import pandas as pd 
import joblib
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from shapash.explainer.smart_explainer import SmartExplainer

import shap
```

```python
# Load the training and testing data
data_split = joblib.load('../data/processed/data_split.joblib')

# Extract split
X_train = data_split['X_train']
X_test = data_split['X_test']
y_train = data_split['y_train']
y_test = data_split['y_test']

# Charger le mod√®le depuis le fichier joblib
loaded_model_data = joblib.load('../models/decision_tree_model.joblib')
loaded_best_model = loaded_model_data['best_model']
```

```python
# Explain the logistic regression model's predictions using SHAP values
explainer = shap.Explainer(loaded_best_model, X_train)
shap_values = explainer(X_test)
```

```python
shap.summary_plot(shap_values, X_test, plot_type="bar", plot_size=(10, 5), show=False)
plt.show()
```

# Shapash

```python
xpl = SmartExplainer(model=loaded_best_model)

# Remove the `model` argument from the compile() method call.
xpl.compile(
    x=X_test
)

app = xpl.run_app()
```
